```python
import json
```

```python
def get_data(filename):
    data = {}
    with open(filename,'r') as file:
        data=json.load(file)
    return data;
```

```python
data = get_data("massive-cleaned.json")
```

```python
def get_friend_suggations(data,ui):
    # all friends to each id
    user_friends = {}
    for user in data["users"]:
        user_friends[user["id"]] = set(user["friends"])
    if(ui not in user_friends):
        return []

    direct_friends = user_friends[ui]

    suggations = {}
    for friend in direct_friends:
        for mutual in user_friends[friend]:
            if(mutual != ui or mutual not in user_friends):
                suggations[mutual] = suggations.get(mutual,0)+1

    sorted_suggations = sorted(suggations.items() ,key =  lambda x:x[1], reverse=True)
            
    return sorted_suggations
```
