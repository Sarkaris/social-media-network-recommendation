```python
def pages_you_may_like(data,ui):
    
    # all liked pages of all users
    all_liked = {}
    for user in data["users"]:
        all_liked[user["id"]] = set(user["liked_pages"])
    if(ui not in all_liked):
        return []

    my_fav = all_liked[ui]
    suggations = {}
    for id,liked_pages in all_liked.items():
        if(id!=ui):
            mutual = my_fav.intersection(liked_pages) 
            other_likes = liked_pages - my_fav 
            value_to_add = suggations.get(id,0)+len(mutual)
            suggations[id] = value_to_add

    # print(all_liked)
    # print(suggations)
            
    # return sorted_suggations
    return sorted(suggations.items(), key = lambda x:x[1] , reverse=True)
```
