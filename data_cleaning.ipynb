```python
import json
```

```python
def get_data(filename):
    data = {}
    with open(filename,'r') as file:
        data=json.load(file)
    return data;
```

```python
data = get_data("massive.json")
```

```python
def display_data(data):
    users = data.get('users')
    for u in users:
        print(f"{u.get('id')} -> name is {u.get('name')} liked pages are {u.get('liked_pages')} and friends are {u.get('friends')}")
    
    pages = data.get('pages')
    print("\n")
    for p in pages:
        print(f"{p.get('id')} -> Title is {p.get('name')} ")
    
    
```

```python
def clean_data(data):
    
    # remove empty names
    data['users']= [user for user in data.get('users') if user.get('name').strip()]
              
    # inactive users removel by friends and likes
    data['users'] = [user for user in data['users'] if user['friends'] or user['liked_pages']]
    

    # Remove duplicate pages
    unique_pages={}
    for page in data['pages']:
        unique_pages[page['id']]=page
    data['pages']=list(unique_pages.values())

    #  remove invalid friends
    # print(valid_users)

    
    # remove duplicate friend ids
    valid_users = [user['id'] for user in data['users'] ]
    for user in data['users']:
        user['friends'] = list(set(fr for fr in user['friends'] if fr in valid_users))
        user['liked_pages'] = list(set(user['liked_pages']))
    # print(valid_users)
    return data;    


clean_data(data)
with open('massive-cleaned.json','w') as file:
    json.dump(data,file)
# print(data)
```
